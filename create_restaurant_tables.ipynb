```python
import re
def find_city(address):
    commaIndex = [w.start() for w in re.finditer(',',address)]
    try:
        city = address[commaIndex[-3]+2:commaIndex[-2]]
    except Exception as e: 
        city = None
    return city

def find_postcode(address):
    commaIndex = [w.start() for w in re.finditer(',',address)]
    try:
        po = address[commaIndex[-2]+5:commaIndex[-1]]
    except Exception as e: 
        po = None
    return po
#      try:
#             num = temp['num_reviews_total'][0]
#         except Exception as e:
#             num = 0

```
