```python
from selenium import webdriver
driver = webdriver.Chrome(executable_path=r'C:\Users\hajih\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
url = 'https://en.wikipedia.org/wiki/Avocado'
avocado_page = driver.get(url)
```

```python
# get review content that are shown on the page at the moment
soup = BeautifulSoup(driver.page_source,'html.parser')
links_list = soup.find_all('a',href= True)
```

```python
# 'avocado' in soup.get_text();
```

```python
links=[]
[links.append(link['href']) for link in links_list]
len(links)
```

```python
keyword = '/wiki/';
wikiLinks = []
# keyword in links[3]
for linkTxt in links:
    if keyword in linkTxt:
        if linkTxt.find(keyword) == 0:
            wikiLinks.append('https://en.wikipedia.org'+linkTxt)
print(len(wikiLinks))
print(len(set(wikiLinks)))
```

```python
import collections
wikiDict = collections.Counter(wikiLinks)
print(len(wikiLinks))
print(len(wikiDict.keys()))
print(len(set(wikiLinks)))
wikiDict
```

```python
# links = []
firstLinks = set(wikiLinks)
for i,url in enumerate(firstLinks):
    driver.get(url);
    soup = BeautifulSoup(driver.page_source,'html.parser')
    if 'avocado' in soup.get_text() or 'Avocado' in soup.get_text():
        for link in soup.find_all('a',href= True):
            temp = link['href']
            if keyword in temp:
                if temp.find(keyword)==0:
                    temp = 'https://en.wikipedia.org'+temp
                    if temp in wikiDict.keys():
                        wikiDict[temp] += 1
                    else:
                        wikiDict[temp] = 1
    if i%20 == 0:
        print(i)

len(wikiDict)
```
