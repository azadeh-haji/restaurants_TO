```python
import pandas as pd
```

```python
rests = pd.read_csv('data/database/restaurants.csv')
len(rests)
(rests.duplicated(subset='place_id')).any()
```

```python
reviews_folder = 'data/reviews/'
places = rests['place_id'].tolist()
rests2 = rests.copy()
plc = []
revid = []
user = []
body = []
rating = []
date_ret = []
date_rel = []
resp  = []
response = []
revcount = []
for i,place in enumerate(places):
    
    #read review file
    file = reviews_folder+place+'.csv'
    df = pd.read_csv(file)
    if len(df) >0 :
        #update review table
        plc      = plc       + [place]*len(df) # append two lists
        revid    = revid     + df['revid'].tolist()
        user     = user      + df['userid'].tolist()
        body     = body      + df['review-text'].tolist()
        rating   = rating    + df['rating'].apply(lambda x: x[1:x.find('star')-1] if x.find('star')>=0 else None).tolist()
        date_ret = date_ret  + df['review-retrieve-date'].tolist()
        date_rel = date_rel  + df['review-relative-date'].tolist()
        resp     = resp      + (df['response-text'].apply(lambda text: False if text == 'None' else True)).tolist()
        response = response  + df['response-text'].tolist()
        
        #update restaurant table
        revcount.append(int(df['num_reviews_total'][0]))
    else:
        revcount.append(0)
        
#         except Exception as e:
#             revcount = None
#         rowInd = rests2['place_id'][rests2['place_id'].eq(place)].index
#         colInd = rests2.columns.get_loc('place_id')
#         rests2.loc[rowInd]['review_count'] = revcount


    print(f'{i} from {len(rests2)}')
```
