```python
cities = ['Toronto','North York','Etobicoke','Scarborough']
for city in cities:
    fig, ax = plt.subplots(figsize=(5,5));
    sns.scatterplot(x = 'fleschEaseReviews', y = 'fleschEaseResponses', data= data[data.city==city],ax=ax)
    plt.title('readability of Response vs Review in '+city)
```
